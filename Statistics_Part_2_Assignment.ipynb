1: What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or inferences about a population based on a sample of data. It helps you determine whether there is enough evidence in your sample to support a specific claim or hypothesis about the population.

2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - It claims there is no effect, no difference, or no relationship between variables.
  - Null Hypothesis:
    - Status quo or no effect
    - Yes, unless proven otherwise
    - p-value ≤ α
  - Alternative hypothesis:
    - New claim or effect
    - No, it’s tested against H₀
    - ...then evidence supports H₁

3: What is the significance level in hypothesis testing, and why is it important?
  - The significance level, denoted by α (alpha), is the threshold used in hypothesis testing to decide whether to reject the null hypothesis (H₀).
  - It is important because:
    - Decision-Making:
      - It sets the rule for deciding whether your p-value is "small enough" to reject H₀.
    - Controls Error Risk:
      - It helps manage the risk of making a Type I error—thinking there's an effect or difference when there actually isn't.

4: What does a P-value represent in hypothesis testing?
  - The p-value is the probability of getting results as extreme as (or more extreme than) the ones you observed assuming the null hypothesis (H₀) is true.

5: How do you interpret the P-value in hypothesis testing?
  - The p-value tells you how likely it is to observe your sample data—or something even more extreme—if the null hypothesis (H₀) is actually true.

6: What are Type 1 and Type 2 errors in hypothesis testing?
  - Type 1 error (False Positive):
    - You reject the null hypothesis (H₀), but it’s actually true.
    - Basically, you’re saying something is happening, when it isn’t.
    - Controlled by the significance level α (commonly 0.05).
  - TYpe 2 error (false Negative):
    - You fail to reject H₀, but H₀ is actually false.
    - You’re saying nothing is happening, when there actually is something going on.
    - Probability of Type II error is denoted by β (beta).

7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
  - One-Tailed Test:
    - One Direction (>, <)
    - One tail of the curve
    - All in one direction
    - μ > 100 or μ < 100
  - two-Tailed Test:
    - Both diection (≠)
    - Both tails of the curve
    - Split across both sides
    - μ ≠ 100

8: What is the Z-test, and when is it used in hypothesis testing?
  - The Z-test is a statistical test used to determine whether there is a significant difference between a sample statistic (like a mean or proportion) and a population parameter, when the population standard deviation is known.
  - The sample size is large (n ≥ 30), or the population is normally distributed.
  - The population standard deviation (σ) is known.
  - You're testing a hypothesis about the population mean or proportion.

9: How do you calculate the Z-score, and what does it represent in hypothesis testing?
  - Z = σ / n X −μ 0
    - X = sample mean
    - 𝜇 0 μ 0 = population mean under H₀
    - 𝜎 σ = population standard deviation.
    - n = sample size
  - It standardizes your sample result on a normal distribution.
  - It shows how far (in standard deviations) your sample is from what you'd expect if H₀ were true.
  
10: What is the T-distribution, and when should it be used instead of the normal distribution?
  - The t-distribution (aka Student’s t-distribution) is a probability distribution used when:
    - You're estimating the mean of a normally distributed population, but the population standard deviation (σ) is unknown.
    - Instead, you're using the sample standard deviation (s) as an estimate.
  - UseCase:-
    - Population standard deviation (σ) is known Z-distribution.
    - σ is unknown and sample size is small (n < 30) T-distribution.
    - σ is unknown but sample size is large (n ≥ 30) T-distribution is still fine (normal approx OK too).

11: What is the difference between a Z-test and a T-test?
  - Z-Test:
    - It is known as population standard deviation
    - It is Typically large than the sample size
    - Standard normal distribution is used here.
    - Degree of freedom is not used.
  - T-Test:
    - Population standard deviation is unknown.
    - Typically small than the sampple size
    - T-distribution is used here.
    - Degree of freedom is used here.

12: What is the T-test, and how is it used in hypothesis testing?
  - A T-test is a statistical test used to determine whether there is a significant difference between:
    - A sample mean and a known population mean.
    - Two sample means.
  - You use a T-test when:
    - You are comparing means.
    - The sample size is small (n < 30) or population standard deviation (σ) is unknown.

13: What is the relationship between Z-test and T-test in hypothesis testing?
  - The relationship between z-test and T-test are:
    - Are used to compare a sample statistic (usually a mean) to a population value or another sample.
    - Help determine whether to reject the null hypothesis (H₀).
    - Use a test statistic (Z or t) calculated from sample data.

14: What is a confidence interval, and how is it used to interpret statistical results?
  - A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence.
  - Estimation: Gives a range of plausible values for the population parameter.
  - Decision-making: If your CI does NOT include the value from the null hypothesis, you may reject H₀.
  - Communicating uncertainty: Shows not just the estimate, but how precise that estimate is.

15: What is the margin of error, and how does it affect the confidence interval?
  - The margin of error is the maximum expected difference between the sample estimate (like a mean or proportion) and the true population value.
  - The confidence interval is built like this:
    - A larger MOE = wider confidence interval = more uncertainty.
    - A smaller MOE = narrower interval = more precision.

16: How is Bayes' Theorem used in statistics, and what is its significance?
  - Bayes’ Theorem describes how to update the probability of a hypothesis based on new evidence.
  - Bayes’ Theorem is the foundation of Bayesian statistics, which differs from traditional (frequentist) stats by:
  - Frequentist Approach:
    -  Long-run frequency
    - Not used
    - Point estimates, confidence intervals.
  - Bayesian Approach:
    - Degree of belief
    - used and updated with data
    - probability distributions for parameters

17: What is the Chi-square distribution, and when is it used?
  - The Chi-square distribution is a right-skewed distribution that arises when you sum the squares of independent standard normal variables.
  - χ 2 = Z 1 2 + Z 2 2 +⋯+Z k 2
  -  When Is the Chi-square Distribution Used:
    - It's used in hypothesis tests and confidence intervals involving variances and categorical data.

18: What is the Chi-square goodness of fit test, and how is it applied?
  - It’s a statistical test used to compare the observed frequencies of outcomes in categories with the expected frequencies based on a specific hypothesis.
  - Steps to apply it:
    - State the hypotheses:
      - H₀: The observed data fits the expected distribution.
    - Calculate expected frequencies:
      - Based on proportions in your hypothesis.
    - Use the formula to compute the Chi-square statistic.

19: What is the F-distribution, and when is it used in hypothesis testing?
  - The F-distribution is a continuous probability distribution that arises when you take the ratio of two independent chi-square variables.
  - 1. ANOVA (Analysis of Variance):
    - Used to compare three or more group means to test if at least one group mean is significantly different.
  - 2. Comparing Two Variances:
    - The F-test can also be used to compare the variances of two populations:

20: What is an ANOVA test, and what are its assumptions?
  - ANOVA stands for Analysis of Variance. It's a statistical test used to determine whether there are significant differences between the means of three or more independent groups.
  - To use ANOVA properly, certain assumptions must be met:
    - Independence of observations:
      - Data from each group must be collected independently.
    - Normality:
      - The data in each group should be approximately normally distributed.
    - Homogeneity of variances:
      - The variances of the groups should be roughly equal.

21: What are the different types of ANOVA tests?
  - One-Way ANOVA:
    - Used to compare the means of three or more independent groups based on a single factor or independent variable.
  - Two-Way ANOVA:
    - Used when you have two independent variables (factors) and you want to examine:
    - The main effect of each factor.
  - Repeated Measures ANOVA:
    - Used when you have repeated measurements on the same subjects under different conditions or over time.
  - Multivariate Analysis of Variance:
    - Used when you have multiple dependent variables that you want to test simultaneously for differences across groups.

22: What is the F-test, and how does it relate to hypothesis testing?
  - The F-test is a statistical test that evaluates whether there is a significant difference between the variances of two or more groups.
  - Comparing Two Variances:
    - The F-test is most commonly used to compare the variances of two independent groups.
  - In ANOVA:
    - In the context of ANOVA, the F-test is used to compare the variance between group means to the variance within the groups.
  - In Regression Analysis:
    - In regression, the F-test is used to determine whether the model as a whole explains a significant portion of the variance in the dependent variable.

```python
# Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?
import math
import scipy.stats as stats


def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    z_statistic = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis (significant difference).")
    else:
        print(f"Fail to reject the null hypothesis (no significant difference).")

sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 25
alpha = 0.05

z_test(sample_mean, population_mean, population_std, sample_size, alpha)
```

```python
# Perform a two-tailed Z-test using Python and visualize the decision region on a plot?
import numpy as np
import math
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample_data, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_statistic = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis (significant difference).")
    else:
        print(f"Fail to reject the null hypothesis (no significant difference).")

    visualize_z_test(z_statistic, alpha)

def visualize_z_test(z_statistic, alpha):
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="Standard Normal Distribution", color='blue')

    critical_value = stats.norm.ppf(1 - alpha / 2)
    plt.fill_between(x, 0, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.3, label='Rejection Region')

    plt.axvline(x=z_statistic, color='green', linestyle='dashed', label=f"Z-statistic: {z_statistic:.2f}")

    plt.axvline(x=-critical_value, color='black', linestyle='dashed', label=f"Critical Value: ±{critical_value:.2f}")
    plt.axvline(x=critical_value, color='black', linestyle='dashed')

    plt.title("Two-Tailed Z-Test Visualization")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend(loc="upper left")
    plt.grid(True)
    plt.show()

np.random.seed(42)
sample_size = 30
population_mean = 100
population_std = 15

sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

two_tailed_z_test(sample_data, population_mean, population_std)
```

```python
# Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(population_mean, population_std, alternative_mean, sample_size, alpha=0.05):

    critical_value = stats.norm.ppf(1 - alpha / 2)


    null_population_data = np.random.normal(population_mean, population_std, 10000)


    alternative_population_data = np.random.normal(alternative_mean, population_std, 10000)


    type_1_error = np.sum(null_population_data > critical_value) / len(null_population_data)


    type_2_error = np.sum(alternative_population_data <= critical_value) / len(alternative_population_data)


    print(f"Type 1 Error (False Positive) Probability: {type_1_error:.4f}")
    print(f"Type 2 Error (False Negative) Probability: {type_2_error:.4f}")


    plot_errors(population_mean, population_std, critical_value, type_1_error, type_2_error)


def plot_errors(population_mean, population_std, critical_value, type_1_error, type_2_error):

    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)

    y = stats.norm.pdf(x, population_mean, population_std)


    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="Normal Distribution (Population)", color='blue')


    plt.fill_between(x, 0, y, where=(x >= critical_value), color='red', alpha=0.3, label="Type 1 Error Region")

    plt.fill_between(x, 0, y, where=(x <= critical_value), color='yellow', alpha=0.3, label="Type 2 Error Region")


    plt.axvline(x=critical_value, color='black', linestyle='dashed', label=f"Critical Value: ±{critical_value:.2f}")

    plt.title("Visualization of Type 1 and Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-Score")
    plt.ylabel("Probability Density")
    plt.legend(loc="upper left")
    plt.grid(True)
    plt.show()

population_mean = 100
population_std = 15
alternative_mean = 105
sample_size = 30
alpha = 0.05

visualize_errors(population_mean, population_std, alternative_mean, sample_size, alpha)
```

```python
# Write a Python program to perform an independent T-test and interpret the results?
import numpy as np
from scipy import stats

def perform_independent_t_test(group1, group2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)

    mean1 = np.mean(group1)
    mean2 = np.mean(group2)

    print(f"Group 1 Mean: {mean1:.4f}")
    print(f"Group 2 Mean: {mean2:.4f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject the null hypothesis (significant difference between means).")
    else:
        print(f"Result: Fail to reject the null hypothesis (no significant difference between means).")

np.random.seed(42)

group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=105, scale=10, size=30)

perform_independent_t_test(group1, group2)
```

```python
# Simulate data and perform both Z-test and T-test, then compare the results using Python?
import numpy as np
import scipy.stats as stats
import math

def simulate_and_compare_tests(population_mean=100, population_std=15, sample_size=25, alpha=0.05):
    np.random.seed(42)
    sample = np.random.normal(loc=population_mean + 5, scale=population_std, size=sample_size)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)

    se_z = population_std / math.sqrt(sample_size)
    z_score = (sample_mean - population_mean) / se_z
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_score)))

    se_t = sample_std / math.sqrt(sample_size)
    t_score = (sample_mean - population_mean) / se_t
    p_value_t = 2 * (1 - stats.t.cdf(abs(t_score), df=sample_size - 1))

    print("=== Sample Statistics ===")
    print(f"Sample Mean        : {sample_mean:.4f}")
    print(f"Sample Std Dev     : {sample_std:.4f}")
    print(f"Sample Size        : {sample_size}")
    print(f"Known Population SD: {population_std}\n")

    print("=== Z-test Result ===")
    print(f"Z-score : {z_score:.4f}")
    print(f"P-value : {p_value_z:.4f}")
    print("Conclusion:", "Reject H₀" if p_value_z < alpha else "Fail to Reject H₀")

    print("\n=== T-test Result ===")
    print(f"T-score : {t_score:.4f}")
    print(f"P-value : {p_value_t:.4f}")
    print("Conclusion:", "Reject H₀" if p_value_t < alpha else "Fail to Reject H₀")

simulate_and_compare_tests()
```

```python
#  Write a Python function to calculate the confidence interval for a sample mean and explain its significance?
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):

    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)

    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean       : {mean:.4f}")
    print(f"Confidence Level  : {confidence*100:.1f}%")
    print(f"Margin of Error   : ±{margin_of_error:.4f}")
    print(f"Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

    return (lower_bound, upper_bound)

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

confidence_interval(sample_data, confidence=0.95)
```

```python
# Write a Python program to calculate the margin of error for a given confidence level using sample data?
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_crit * standard_error

    print(f"Sample Size         : {n}")
    print(f"Sample Std Dev      : {sample_std:.4f}")
    print(f"Standard Error      : {standard_error:.4f}")
    print(f"Confidence Level    : {confidence*100:.1f}%")
    print(f"T-critical value    : {t_crit:.4f}")
    print(f"Margin of Error     : ±{margin_of_error:.4f}")

    return margin_of_error

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

calculate_margin_of_error(sample_data, confidence=0.95)

```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?
def bayesian_inference(prior, sensitivity, false_positive_rate):
    p_positive = sensitivity * prior + false_positive_rate * (1 - prior)

    posterior = (sensitivity * prior) / p_positive

    print("=== Bayesian Inference ===")
    print(f"Prior (P(Disease))           : {prior:.4f}")
    print(f"Sensitivity (P(+|Disease))   : {sensitivity:.4f}")
    print(f"False Positive Rate (P(+|~D)): {false_positive_rate:.4f}")
    print(f"Marginal P(Positive Test)    : {p_positive:.4f}")
    print(f"Posterior P(Disease|Positive): {posterior:.4f}")

    return posterior

prior = 0.01
sensitivity = 0.99
false_positive_rate = 0.05


bayesian_inference(prior, sensitivity, false_positive_rate)
```

```python
# Perform a Chi-square test for independence between two categorical variables in Python?
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency


data = np.array([[30, 20],
                 [25, 25]])

df = pd.DataFrame(data, columns=["Like", "Dislike"], index=["Male", "Female"])

print("Contingency Table:")
print(df)

chi2, p, dof, expected = chi2_contingency(data)

print("\nChi-square Test Results:")
print(f"Chi2 Statistic : {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value         : {p:.4f}")
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, columns=["Like", "Dislike"], index=["Male", "Female"]))


alpha = 0.05
if p < alpha:
    print("\nConclusion: Reject the null hypothesis (variables are dependent).")
else:
    print("\nConclusion: Fail to reject the null hypothesis (variables are independent).")
```

```python
# Write a Python program to calculate the expected frequencies for a Chi-square test based on observed  data?
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    observed = np.array(observed)

    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    grand_total = observed.sum()

    expected = (row_totals @ col_totals) / grand_total
    return expected


observed_data = np.array([
    [30, 20],
    [25, 25]
])

expected_frequencies = calculate_expected_frequencies(observed_data)

observed_df = pd.DataFrame(observed_data, columns=["Like", "Dislike"], index=["Male", "Female"])
expected_df = pd.DataFrame(expected_frequencies, columns=["Like", "Dislike"], index=["Male", "Female"])

print("Observed Frequencies:")
print(observed_df)
print("\nExpected Frequencies (if variables are independent):")
print(expected_df.round(2))
```

```python
# Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution?
import numpy as np
from scipy.stats import chisquare

observed = np.array([8, 9, 10, 12, 11, 10])

expected = np.array([10, 10, 10, 10, 10, 10])

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("=== Chi-square Goodness-of-Fit Test ===")
print(f"Observed Frequencies : {observed}")
print(f"Expected Frequencies : {expected}")
print(f"Chi2 Statistic        : {chi2_stat:.4f}")
print(f"P-value               : {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis (data does NOT fit expected distribution).")
else:
    print("Conclusion: Fail to reject the null hypothesis (data fits expected distribution).")
```

```python
# Implement an F-test using Python to compare the variances of two random samples?
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)


    f_stat = var1 / var2 if var1 >= var2 else var2 / var1


    df1 = len(sample1) - 1
    df2 = len(sample2) - 1


    p_value = 1 - f.cdf(f_stat, df1, df2)


    print("=== F-test for Equality of Variances ===")
    print(f"Sample 1 Variance: {var1:.4f}")
    print(f"Sample 2 Variance: {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"P-value: {p_value:.4f}")


    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (variances are significantly different).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (variances are equal).")


np.random.seed(42)
sample1 = np.random.normal(0, 10, 50)
sample2 = np.random.normal(0, 12, 50)

f_test(sample1, sample2, alpha=0.05)
```

```python
#  Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results?
import numpy as np
from scipy.stats import f_oneway

def perform_anova(*groups):
    f_statistic, p_value = f_oneway(*groups)

    print("=== ANOVA Test Results ===")
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (there is a significant difference in means).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (means are not significantly different).")

np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

perform_anova(group1, group2, group3)
```

```python
# Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway


def perform_one_way_anova(*groups):

    f_statistic, p_value = f_oneway(*groups)


    print("=== One-Way ANOVA Test Results ===")
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")


    alpha = 0.05
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (there is a significant difference in means).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (means are not significantly different).")


np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)


perform_one_way_anova(group1, group2, group3)


def plot_boxplots(*groups):

    plt.figure(figsize=(8, 6))
    plt.boxplot(groups, labels=['Group 1', 'Group 2', 'Group 3'])
    plt.title("One-Way ANOVA: Comparison of Group Means")
    plt.ylabel("Values")
    plt.show()

plot_boxplots(group1, group2, group3)
```

```python
#  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA?
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene
import seaborn as sns

def check_assumptions(*groups):
    """
    Check assumptions for ANOVA:
    1. Normality (Shapiro-Wilk test)
    2. Homogeneity of variance (Levene's test)
    """

    print("=== Normality Check (Shapiro-Wilk Test) ===")
    for i, group in enumerate(groups, 1):
        stat, p_value = shapiro(group)
        print(f"Group {i}: Shapiro-Wilk Test p-value = {p_value:.4f}")
        if p_value < 0.05:
            print("Conclusion: Data is not normally distributed in this group.\n")
        else:
            print("Conclusion: Data is normally distributed in this group.\n")

    print("=== Homogeneity of Variance Check (Levene's Test) ===")
    stat, p_value = levene(*groups)
    print(f"Levene's Test p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("Conclusion: Variances are significantly different (violates homogeneity of variances assumption).\n")
    else:
        print("Conclusion: Variances are equal (assumption satisfied).\n")


    print("=== Visualizations ===")

    plt.figure(figsize=(12, 6))
    for i, group in enumerate(groups, 1):
        plt.subplot(1, len(groups), i)
        sns.histplot(group, kde=True)
        plt.title(f"Group {i} Histogram")
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(8, 6))
    plt.boxplot(groups, labels=[f'Group {i}' for i in range(1, len(groups) + 1)])
    plt.title("Box Plot: Homogeneity of Variances Check")
    plt.ylabel("Values")
    plt.show()

np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 15, 30)


check_assumptions(group1, group2, group3)
```

```python
# Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the  results?
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy.stats import f
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(42)

factor_A = np.random.choice(['A', 'B', 'C'], size=90)

factor_B = np.random.choice(['Male', 'Female'], size=90)

scores = np.random.normal(loc=50, scale=10, size=90)


interaction_effect = (factor_A == 'A') * (factor_B == 'Male') * 5
scores += interaction_effect

data = pd.DataFrame({'Factor_A': factor_A, 'Factor_B': factor_B, 'Scores': scores})

model = ols('Scores ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("=== Two-Way ANOVA Results ===")
print(anova_table)

alpha = 0.05
if anova_table['PR(>F)']['C(Factor_A)'] < alpha:
    print("Factor A has a significant effect on scores.")
else:
    print("Factor A does not have a significant effect on scores.")

if anova_table['PR(>F)']['C(Factor_B)'] < alpha:
    print("Factor B has a significant effect on scores.")
else:
    print("Factor B does not have a significant effect on scores.")

if anova_table['PR(>F)']['C(Factor_A):C(Factor_B)'] < alpha:
    print("There is a significant interaction effect between Factor A and Factor B.")
else:
    print("There is no significant interaction effect between Factor A and Factor B.")

plt.figure(figsize=(10, 6))
sns.pointplot(x="Factor_A", y="Scores", hue="Factor_B", data=data, markers=["o", "s"], linestyles=["-", "--"])
plt.title("Two-Way Interaction Plot")
plt.xlabel("Factor A (Teaching Method)")
plt.ylabel("Test Scores")
plt.legend(title="Factor B (Gender)")
plt.show()
```

```python
# Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means?
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)

group_A = np.random.normal(50, 10, 30)
group_B = np.random.normal(55, 10, 30)
group_C = np.random.normal(60, 15, 30)

f_stat, p_value = stats.f_oneway(group_A, group_B, group_C)

print("=== One-Way ANOVA Results ===")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is a significant difference between the group means.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference between the group means.")


plt.figure(figsize=(8, 6))
sns.boxplot(data=[group_A, group_B, group_C], notch=True, patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='black'),
            whiskerprops=dict(color='black'), flierprops=dict(markerfacecolor='red', marker='o', markersize=8))

plt.title("One-Way ANOVA: Comparing Group Means", fontsize=14)
plt.xticks([0, 1, 2], ['Group A', 'Group B', 'Group C'], fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.show()
```

```python
# Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means?
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(42)

population_mean = 100
population_std = 15
sample_size = 50


sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)


sample_mean = np.mean(sample_data)
t_stat, p_value = stats.ttest_1samp(sample_data, population_mean)


alpha = 0.05

print("=== Hypothesis Testing Results ===")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Conclusion: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")


plt.figure(figsize=(8, 6))
plt.hist(sample_data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f'Sample Mean = {sample_mean:.2f}')
plt.axvline(population_mean, color='green', linestyle='dashed', linewidth=2, label=f'Population Mean = {population_mean}')
plt.title("Histogram of Simulated Data with Sample Mean and Population Mean", fontsize=14)
plt.xlabel("Data Values", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.legend()
plt.show()
```

```python
# Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results?
import numpy as np
import scipy.stats as stats

np.random.seed(42)
sample_size = 30
population_variance = 25
sample_data = np.random.normal(50, np.sqrt(population_variance), sample_size)

sample_variance = np.var(sample_data, ddof=1)

hypothesized_variance = 25

n = sample_size
test_statistic = (n - 1) * sample_variance / hypothesized_variance

alpha = 0.05
df = n - 1
critical_value_lower = stats.chi2.ppf(alpha / 2, df)
critical_value_upper = stats.chi2.ppf(1 - alpha / 2, df)

p_value_lower = stats.chi2.cdf(test_statistic, df)
p_value_upper = 1 - p_value_lower
p_value = min(p_value_lower, p_value_upper) * 2

print("=== Chi-Square Test for Population Variance ===")
print(f"Sample variance: {sample_variance:.4f}")
print(f"Test statistic: {test_statistic:.4f}")
print(f"Critical value (lower): {critical_value_lower:.4f}")
print(f"Critical value (upper): {critical_value_upper:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The population variance is significantly different from the hypothesized value.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference between the sample variance and the hypothesized population variance.")
```

```python
# Write a Python script to perform a Z-test for comparing proportions between two datasets or groups?
import numpy as np
import scipy.stats as stats

x1 = 80
n1 = 200
x2 = 70
n2 = 180

p1_hat = x1 / n1
p2_hat = x2 / n2

P = (x1 + x2) / (n1 + n2)

Z = (p1_hat - p2_hat) / np.sqrt(P * (1 - P) * (1/n1 + 1/n2))

p_value = 2 * (1 - stats.norm.cdf(abs(Z)))

print("=== Z-test for Comparing Proportions ===")
print(f"Proportion of successes in group 1: {p1_hat:.4f}")
print(f"Proportion of successes in group 2: {p2_hat:.4f}")
print(f"Pooled proportion: {P:.4f}")
print(f"Z-statistic: {Z:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. There is a significant difference between the proportions.")
else:
    print("Conclusion: Fail to reject the null hypothesis. There is no significant difference between the proportions.")
```

```python
# Perform a Chi-square test for goodness of fit with simulated data and analyze the results.?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)
observed_data = np.random.choice([1, 2, 3, 4, 5, 6], size=100, p=[0.1, 0.2, 0.2, 0.1, 0.2, 0.2])

observed_freq = np.array([np.sum(observed_data == i) for i in range(1, 7)])


expected_prob = [0.1, 0.2, 0.2, 0.1, 0.2, 0.2]
total_rolls = len(observed_data)
expected_freq = np.array(expected_prob) * total_rolls

chi_square_statistic, p_value = stats.chisquare(observed_freq, expected_freq)

print("=== Chi-square Goodness of Fit Test ===")
print(f"Observed Frequencies: {observed_freq}")
print(f"Expected Frequencies: {expected_freq}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the null hypothesis. The observed frequencies do not match the expected frequencies.")
else:
    print("Conclusion: Fail to reject the null hypothesis. The observed frequencies match the expected frequencies.")

labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(x - 0.2, observed_freq, 0.4, label='Observed', color='blue')
ax.bar(x + 0.2, expected_freq, 0.4, label='Expected', color='red', alpha=0.7)
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_xlabel('Dice Faces')
ax.set_ylabel('Frequency')
ax.set_title('Chi-square Goodness of Fit: Observed vs Expected Frequencies')
ax.legend()

plt.show()
```
